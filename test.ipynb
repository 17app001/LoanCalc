```python
loanAmount=100000
interestRate=5
loanYear=5
```

# 本金攤還

```python
proid=loanYear*12
proid
```

### 每個月償還本金(固定)

```python
monthReplacement=int(loanAmount/proid)
monthReplacement
```

### 每個月償還利息(遞減)
- 貸款餘額*(月利率)

```python
interest=round(loanAmount*(interestRate/100/12))
interest
```

### 計算每個月還款金額 

```python
loanAmount=100000
for i in range(proid):
    interest=round(loanAmount*(interestRate/100/12))

    if i==proid-1:
        print(f'期別:{i+1:3} 償還本金:{monthReplacement+loanAmount-monthReplacement:5} 償還利息:{interest:5}\
 償還本利和:{monthReplacement+loanAmount-monthReplacement+interest:5} 貸款餘額:{0:5}')
    else:
        print(f'期別:{i+1:3} 償還本金:{monthReplacement:5} 償還利息:{interest:5}\
 償還本利和:{monthReplacement+interest:5} 貸款餘額:{loanAmount-monthReplacement:5}')
        
    print('---------------------------------------------------------------------')
    loanAmount-=monthReplacement
```
